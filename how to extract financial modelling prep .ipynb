```python

def get_statement(company_ticker, statement_name, api_key, frequency='annual', df=False):  # df= false stop the default set up of returning a pd df
	
 """
	Get a financial statement to use for fundamental calculations

	inputs:
	company_ticker (str) -- e.g. 'AAPL' for Apple inc.
	statement_name (str) -- one of: 'income-statement','balance-sheet-statement','cash-flow-statement','enterprise-value'
	period (str) -- 'annual' or 'quarter'
	api_key (str) -- api key to access financialmodelingprep account

	returns:
	Pandas DataFrame object
	"""
	
 if statement_name in ['income-statement', 'balance-shee-statement','cash-flow-statement', 'eneterprise-value', 'financial-ratios' ]: 
     # check if the statement name is valid and or available for the given company
     response= requests.get(f'https://financialmodelingprep.com/api/v3/{statement_name}/{company_ticker}?period={frequency}&apikey={api_key}')
     # website name + the variable inside our function then converted into j.son which convert the API responde into a python dictionary list
     
     
     if response.status_code == 200: 
         statement = response.json() # only fetching correct dat if status code == 401 and so on ther eis an error in the code or data are not there 
         
         
         if df: #convert to pandas 
             statement = pd.DataFrame.from_dict(statement) # what does statemetn [0] do?
             #statement = statement.iloc[5: -2] # why are we removing the first 5 and last 2 rows? 
             #statement.columns= ['Year 0']
             #statement['Year 0'] = statement['Year 0'].astype(float) # what is about year 0? 
             return statement
         else:
             print('Data is not available for this company') 
             return None
             
         
     


```

```python
def df_summary(df, name="DataFrame"):
    print(f"--- {name}.info() ---")
    print(df.info())
    print()
    
    print(f"--- {name}.describe() ---")
    print(df.describe(include='all'))
    print()
    
    print(f"--- {name}.head() ---")
    print(df.head())
    print()
```

```python
# first 5 and last 2 columns are not needed
def remove_col(df):
    # Drop first 6 columns (0 to 5) and last 2 columns (-2, -1)
    cols_to_drop = list(df.columns[:6]) + list(df.columns[-2:])
    return df.drop(columns=cols_to_drop)


```

```python
def remove_zero_columns(df):
    """
    Remove columns with all zero values from the DataFrame.
    
    Parameters:
    df (DataFrame): The DataFrame to process.
    
    Returns:
    DataFrame: The DataFrame with zero-value columns removed.
    """
    
    
for column in df.columns: # removed period & calendar year as they are object types and have full data
    if column in [ 'revenue', 'costOfRevenue', 'grossProfit',
       'grossProfitRatio', 'researchAndDevelopmentExpenses',
       'generalAndAdministrativeExpenses', 'sellingAndMarketingExpenses',
       'sellingGeneralAndAdministrativeExpenses', 'otherExpenses',
       'operatingExpenses', 'costAndExpenses', 'interestIncome',
       'interestExpense', 'depreciationAndAmortization', 'ebitda',
       'ebitdaratio', 'operatingIncome', 'operatingIncomeRatio',
       'totalOtherIncomeExpensesNet', 'incomeBeforeTax',
       'incomeBeforeTaxRatio', 'incomeTaxExpense', 'netIncome',
       'netIncomeRatio', 'eps', 'epsdiluted', 'weightedAverageShsOut',
       'weightedAverageShsOutDil'] :
         if df[column].sum() ==0:
           print(f'{column} only has zero values')
           df.drop(column, axis=1, inplace = True) 
```
